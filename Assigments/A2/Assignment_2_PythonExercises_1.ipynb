 # **Text Analysis Tool** ðŸ“‘

```python
#Functions to use
def text_file_reading(file_path):
    #The content of the text file is goint to be read and it'll be returned as a string

    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
    return text

def count_words(text):
    #The number of word will be counted.

    words = re.findall(r'\b\w+\b', text)
    return len(words)

def count_sentences(text):
    #The number of sentences will be counted.

    sentences = re.split(r'[.!?]', text) # Split the input text wherever it finds a .!?
    sentences = [sentence for sentence in sentences if sentence.strip()] # Removes leading and trailing whitespaces from each sentence
    return len(sentences)

def count_paragraphs(text):
    #The number of paragraphs will be counted.

    paragraphs = text.split('\n\n')  # Splits the input text wherever it finds two consecutive newline characters signifying the start of a new paragraph.
    paragraphs = [paragraph for paragraph in paragraphs if paragraph.strip()]
    return len(paragraphs)

def calculate_average_word_length(text):
    #The average word length will be counted.

    words = re.findall(r'\b\w+\b', text) #The entire pattern \b\w+\b ensures that only complete words are matched.
    total_length = sum(len(word) for word in words) # Calculate the total length of all words in the "words" list.
    return total_length / len(words) if words else 0

def identify_most_common_words(text, n=10):
    #The most common words and their frequencies will be counted.

    words = re.findall(r'\b\w+\b', text.lower())
    word_counts = Counter(words)
    return word_counts.most_common(n)
```

```python
# Call the functions in the main function
def text_analysis_tool():
    file_path = 'Text.txt'  # Give the text file
    text = text_file_reading(file_path)

    num_words = count_words(text)
    num_sentences = count_sentences(text)
    num_paragraphs = count_paragraphs(text)
    average_word_length = calculate_average_word_length(text)
    most_common_words = identify_most_common_words(text)

    print ("So you need help with your analysis text? Don't worry, we got you!")
    print(f"The total number of words is: {num_words} words.")
    print(f"The total number of sentences is: {num_sentences} sentences.")
    print(f"The total number of paragraphs is: {num_paragraphs} paragraphs.")
    print(f"The average word length is: {average_word_length:.2f}")
    print("The most common words are:")
    for word, frequency in most_common_words:
        print(f"{word}: {frequency}")
```

```python
#Run main function
if __name__ == "__main__":
    text_analysis_tool()
```

# **Hangman Game** ðŸ¤¯

```python
#Functions to use
def choose_word():
    # List of possibles words in the game
    word_list = ["red", "blue", "purple", "black", "white", "orange", "pink", "yellow"]
    return random.choice(word_list)

def line_word(word, guessed_letters):
    # This representation helps to show the player which letters they have guessed correctly and which letters are still unknown.
    display = ""
    for letter in word:
        if letter in guessed_letters:
            display += letter
        else:
            display += "_"
    return display

def hangman_drawing(attempts):
    # Phases of the hangman):

    stages = [
        """
           --------
           |      |
           |
           |
           |
           |
        """,
        """
           --------
           |      |
           |      O
           |
           |
           |
        """,
        """
           --------
           |      |
           |      O
           |      |
           |
           |
        """,
        """
           --------
           |      |
           |      O
           |     /|
           |
           |
        """,
        """
           --------
           |      |
           |      O
           |     /|\\
           |
           |
        """,
        """
           --------
           |      |
           |      O
           |     /|\\
           |     /
           |
        """,
        """
           --------
           |      |
           |      x
           |     /|\\
           |     / \\
           |
        """
    ]
    return stages[attempts]
```

```python
#Define the main function
def Hangman():
    print("Welcome to color's hangman game!")
    print("Find out the color!")
    word = choose_word()
    guessed_letters = []
    attempts = 6

    while True:
        print(f"\nYou have {attempts} attempts.")
        print(hangman_drawing(6 - attempts))
        print("The color is: ", line_word(word, guessed_letters))

        guess = input("Guess a letter: ").lower()

        if guess in guessed_letters:
            print("You've already tried that letter):, try again! :)")
            continue

        guessed_letters.append(guess)

        if guess in word:
            print("You have guessed a letter!")
            if all(letter in guessed_letters for letter in word):
                print("You guessed the word!")
                print ("The word was: ", word)
                print ("Take a candy as your prize!")
                break
        else:
            print("That's not a letter):")
            attempts -= 1
            if attempts == 0:
                print(hangman_drawing(6 - attempts))
                print("Sorry, you don't have more attempts): The word was:", word)
                break
```

```python
#Run main function
if __name__ == "__main__":
    Hangman()
```

# **Number Guessing Game** ðŸ”¢

```python
#main function
def guess_the_number():
    # Get a random number from 100 to 500.
    target_number = random.randint(100, 500)

    guessed = False

    print("Do you think you can read my elecronic mind? Find out in the Guess the Number game!")
    print("The number I'm thinking on is between 100 and 500. Good luck!")

    while not guessed:
        # Ask for a number
        guess = int(input("What number I'm thinking about?: "))

        # Check if the guess is correct
        if guess == target_number:
            print("Oh no! You guessed the number correctly. I think humans already use their 6th sense.")
            guessed = True
        # Provide hints if the guess is not correct
        elif guess < target_number:
            print("Humans are still fools, i'll give you a clue: the number is greater.")
        else:
            print("Humans are still fools, i'll give you a clue: the number is less.")
```

```python
# Run the game
if __name__ == "__main__":
    guess_the_number()
```
